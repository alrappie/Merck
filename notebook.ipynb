```python
import pandas as pd
```

```python
data = pd.read_excel('raw_data.xlsx',skiprows=7,sheet_name='Exercise')
data
```

```python
df = data.iloc[:,0].str.split(';', expand=True)
df
```

```python
df.columns = data.columns.str.split(';')[0] + ['Quantity']
df
```

```python
df['Date / Tanggal'] = pd.to_datetime(df['Date / Tanggal']).dt.strftime('%d/%m/%Y')
df
```

```python
data2 = pd.read_excel('raw_data.xlsx',sheet_name='Harga Jual Product')
data2
```

```python
joined_data = df.join(data2.set_index('Product Code / Kode Barang'), on='Product Code / Kode Barang')
joined_data
```

```python
joined_data['Total Harga Jual'] = joined_data['Quantity'].astype('int32') * joined_data['Harga Jual'].astype('int32')
joined_data
```
